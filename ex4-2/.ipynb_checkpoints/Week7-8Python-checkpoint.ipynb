```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime_rate = pd.read_csv('crimerates-by-state-2005.csv')
crime_rate2 = crime_rate[(crime_rate['state'] != 'United States') & (crime_rate['state'] != 'District of Columbia')]
crime_rate
```

```python
plt.scatter(crime_rate2['population'], crime_rate2['murder'])
```

```python
plt.scatter(crime_rate2['population'], crime_rate2['forcible_rape'], s= crime_rate2['murder'] * 40, alpha= .5)
```

```python
crime_rate.murder.plot.density(label= 'Murder')
crime_rate.forcible_rape.plot.density(label= 'Rape')
plt.legend()
```
